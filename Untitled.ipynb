```python
import numpy as np

H = 24


price = np.array([1] * 5 + [6] * 3 + [2]*8 + [9]*3 + [1] * 5)

max_power =  np.array([.2] * 5 + [.05] * 3 + [.2]*8 + [.1]*3 + [.1] * 5)
#max_power = np.array([10]*H)
#max_power = np.array([7] * 24)

class Load:
    profile : np.array

    def __init__(self, profile):
        self.profile = profile
    
    @property
    def duration(self):
        return len(self.profile)
    
class ConstantLoad(Load):
    power = 0
    def __init__(self, power, duration):
        self.power = power
        self.profile = power*np.ones(duration)


loads = []
for l in range(3):
    for d in range(2,14):
        loads.append(ConstantLoad( 0.01 + np.random.uniform(-0.01, 0.01), d ))
    

import pyomo.environ as pyo

model = pyo.ConcreteModel()


def initialize_prices(model, l, t):
    load = loads[l]

    if t > 23 - load.duration:
        return infinity
    else:
        return price[t:t+load.duration].dot(load.profile)

# Indices
model.l = pyo.RangeSet(0,len(loads)-1)
model.t = pyo.RangeSet(0, H-1)

deltas = []

# Parameters
def initialize_deltas(model, t):
    deltas = []
    for l in range(len(loads)):
        for d in range(0, min(loads[l].duration, t+1)):
            deltas.append( (d, l) )

    return deltas

model.deltas = pyo.Param(model.t, initialize = initialize_deltas)

model.prices = pyo.Param(model.l, model.t, initialize=initialize_prices)
model.max_power= pyo.Param(model.t, initialize=max_power)


# Variables
model.start = pyo.Var(model.l, model.t, domain=pyo.Binary)

for l in range(len(loads)):
    for d in range( H - loads[l].duration, H):
        model.start[l,d].fixed = True
        model.start[l,d].value = 0.0

    if loads[l].duration == H:
        model.start[l, 0].value = 1.0

# Constraints
def constraint_activate_once_func(model, l):
    return sum(model.start[l, t] for t in model.t ) == 1

model.constraint_activate_once = pyo.Constraint(model.l, rule=constraint_activate_once_func)


def constraint_max_power(model, t):
    index = model.deltas[t]
    return sum(
               model.start[l,t - d] * loads[l].profile[d]
                   for (d,l) in index
              ) <= model.max_power[t]
        
        
model.constraint_max_power = pyo.Constraint(model.t, rule=constraint_max_power)


# Objective function
def objective_func(model):
    return sum(model.start[l,t] * model.prices[l,t] for l in model.l for t in model.t)
    
model.objective = pyo.Objective(rule=objective_func, sense=pyo.minimize)

#model.pprint()
```

```python
opt = pyo.SolverFactory('cbc')
opt.options = { 'sec': 1000,  'threads':10, 'ratio': 0.02}
result = opt.solve(model)
```
